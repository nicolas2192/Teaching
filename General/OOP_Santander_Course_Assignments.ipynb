- Create a player class with 1 class attribute (MAX_HEARTS) and 2 instance attributes, name and hearts.
- Create a instance method lose() that substracts one heart every time it is called and if the count reaches 0 display 'game over'.

```python
class Player(object):
  'Player Class'

  MAX_HEARTS = 3

  def __init__(self, name):
    self.name = name
    self.hearts = Player.MAX_HEARTS

  def __repr__(self):
    return f'{self.name}, {self.hearts}'

  def lose(self):
    if self.hearts == 1:
      print("You lost. Game Over!")
    else:
      self.hearts -= 1
      print(f'You lost one heart. {self.hearts} remaining.')

p1 = Player('Nicolas')
print(p1)
print(p1.hearts)

p1.lose()
p1.lose()
p1.lose()
```

Create a ShoppingList Class that creates an instance from a list of element and from a string (factory method).

```python
class ShoppingList(object):

  def __init__(self, items: list):
    self.items = items

  def __repr__(self):
    return f'{self.items}'

  @classmethod
  def from_string(cls, string):
    lst = string.split(',')
    return cls(lst)


lst1 = ShoppingList(['Banana', 'Milk', 'Ice Cream'])
print(lst1)

lst2 = ShoppingList.from_string("Banana,Milk,Ice Cream,Chocolate")
print(lst2)
```

Create a Weather Class that accepts two parameters; Status and Temperature.

Status is a string that can take sunny, rainy, cloudy, windy, or snowy as a value, while the temperature is a decimal number ranging from -40 up to 50 centigrades. Implement getters and setters to allow the user to update the attributes.

```python
class Weather():
    STATUS = ['sunny', 'rainy', 'cloudy', 'windy', 'snowy']
    MIN_TEMP = -40
    MAX_TEMP = 50

    def __init__(self, status, temp):
        self.__status = status
        self.__temp = temp

    def __repr__(self):
        return f'{self.__class__.__name__}: {self.__status}, {self.__temp}'

    @property
    def temp(self):
        return self.__temp

    @temp.setter
    def temp(self, new_temp):
        assert 50 >= new_temp >= -40, f'Temperature value should range between {Weather.MIN_TEMP} and {Weather.MAX_TEMP}.'
        self.__temp = new_temp

    @property
    def status(self):
        return self.__status

    @status.setter
    def status(self, new_status):
        assert new_status in Weather.STATUS, f'Status should be one of the following: {Weather.STATUS}'
        self.__status = new_status
```

```python
w1 = Weather("sunny", 28)
w1.temp = 10
w1.status = 'cloudy'
w1
```

CaesarCipher class. The class takes one parameter, an integer number that sets the encryption algorithm. Implement the _ call _ method and pass the text that should be encrypted/decrypted as a parameter.

There should be a method for encrypting, decrypting, and mapping (showing the mapping for the letters).

[Pythom map() function interesting read](https://realpython.com/python-map-function/). Loop without creating loops.

```python
import string

class CaesarCipher():
    'Caesar Cipher'
    MIN_ENCR_VALUE = -20
    MAX_ENCR_VALUE = 20
    ABC = string.ascii_uppercase

    def __init__(self, encr_value: int):
        assert isinstance(encr_value, int), 'The encryption value should be an integer'
        assert CaesarCipher.MIN_ENCR_VALUE <= encr_value <= CaesarCipher.MAX_ENCR_VALUE, f'The encryption value should be an integer between {CaesarCipher.MIN_ENCR_VALUE} and {CaesarCipher.MAX_ENCR_VALUE}'
        self.encr_value = encr_value
        self.encr_key = CaesarCipher.ABC[encr_value:] + CaesarCipher.ABC[:encr_value]
        self.mapping = list(zip(self.ABC, self.encr_key))

    def __repr__(self):
        return f'{self.__class__.__name__}: Encription Value: {self.encr_value}'

    def __encoder(self, letter, mode):
        'Returns the encoded letter' # A dictnionary and using get() might have been better option, instead of a list
        if mode == 'conceal':
            try:
                pos = self.ABC.index(letter)
                return self.encr_key[pos]
            except ValueError as v0:
                return letter
        elif mode == 'reveal':
            try:
                pos = self.encr_key.index(letter)
                return self.ABC[pos]
            except ValueError as v0:
                return letter

    def encrypt(self, text: str):
        'Applies the encoder function to all letters returning the encoded text'
        encrypted = "".join([self.__encoder(let, mode='conceal') for let in text.upper()])
        return encrypted

    def __call__(self, text):
        return self.encrypt(text)

    def decrypt(self, text: str):
        decrypted = "".join([self.__encoder(let, mode='reveal') for let in text.upper()])
        return decrypted
```

Create a Library class that stores Book class objects in it. You should be able to iterate over Library printing all available books. The Library class has on a single method for adding books.

Book class has two parameter, the book's name and category.

```python
class Book():
    def __init__(self, name, category):
        self.name = name
        self.category = category

    def __repr__(self):
        return f"Book: {self.name}, {self.category}"


class Library():
    def __init__(self, name):
        self.name = name
        self.books = []

    def __len__(self):
        return len(self.books)

    def __getitem__(self, indx):
        return self.books[indx]

    def add(self, book: Book):
        assert isinstance(book, Book), f"{book} has to be an instance of the Book class"
        self.books.append(book)
        print('Book added successfully!')
```

```python
b1 = Book('Azul', 'Novel')
b2 = Book('Conejos y Ratones', 'Novel')
b3 = Book('El Color', 'Educational')

print(b1)

l1 = Library('Babilonia')
len(l1)
l1.add(b1)
l1.add(b2)
l1.add(b3)
len(l1)

print(l1[0], l1[1], l1[2])
```

Create a Rectangle class and implement comparable operators.
The Rectangle class takes two arguments, height and width. Internally it computes the instance area.
Operators should be able to:
- Compare two rectangles' area.
- Compare one rectangle's are with a scalar.
- Raise an error if the other object is not a Rectangle nor an integer.

```python
class Rectangle():
    def __init__(self, height, width):
        self.height = height
        self.width = width
        self.area = height * width

    def __repr__(self):
        return f'Rectangle HW: {self.height * self.width}'

    def __gt__(self, other):
        if isinstance(other, Rectangle):
            return self.area > other.area
        elif isinstance(other, int):
            return self.area > other
        else:
            raise ValueError(f'Cannot compare Rectangle and {other.__class__.__name__}')

    def __lt__(self, other):
        if isinstance(other, Rectangle):
            return self.area < other.area
        elif isinstance(other, int):
            return self.area < other
        else:
            raise ValueError(f'Cannot compare Rectangle and {other.__class__.__name__}')
```

```python
r1 = Rectangle(5,4)
r2 = Rectangle(3,6)

print(r1, r2)
print('r1 > r2', r1 > r2)
print('r1 > 30', r1 > 30)
#print(30 > r1) # this raises an error
print('r2 < 99', r2 < 99)
```
