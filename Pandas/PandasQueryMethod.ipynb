```
import pandas as pd
import numpy as np
```

### TL;DR

The .query() command filters data in a similar fashion to loc[], however, it is easier to read and write since its like writing a SQL where statement.

```
df = pd.DataFrame(data=np.random.randint(low=0, high=5, size=(10,4)), columns=['a', 'b', 'c', 'd'])
df.head(3)
```

```
df.query('a > 1')
```

```
df.query('(a > 1) & (b == 0)')
```

```
df.query('a > b')
```

```
cutoff = 3
df.query('d > @cutoff')
```

```
other = df.query('a == 1 | b == 1 | c == 1 | d == 1').copy()
other
```

```
full = df.query('a >= 0')
df is full
```
